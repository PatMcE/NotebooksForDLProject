## Turn Map Tile into patches

Turn into patches:

```python
import numpy as np
import PIL
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open('Ballinameen_7cm.tif')
np_img = np.array(img) #convert the image to a numpy array
np_img = np_img[:,:,:3]

my_list = [[5500,1000],
          [6500,1000],
           [7500,1000],
           [8300,1000],
           [7000,2000],
           [5500,2500],
           [7500,2500],
           [8500,2500],
           [500,4000],
           [2500,4000],
           [4500,4000],
           [4000,4600],
           [6500,4000],
           [8000,4000],
           [2000,4600],
           [1500,5000],
           [3500,5000],
           [5500,5000],
           [6300,5000],
           [7000,3000]   
          ]

id = 1
for coords in my_list:
    arr = np_img[coords[1]:coords[1]+512, coords[0]:coords[0]+512, :] #patches 512px by 512px
    img = Image.fromarray(arr)
    img.save(f"new_images/sq{id}.png") #save all patches in folder 'new_images'
    id+=1
```

Plot Map Tile:

```python
plt.imshow(np_img)
```

Plot Example Patch:

```python
def plot_a_patch(np_img, x,y):
    plt.imshow(np_img[y:y+512, x:x+512, :])
```

```python
plot_a_patch(np_img, 7000,3000)
```
