# Plot Bar Charts for Conference Paper

## Imports:

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (7,5))
```

## Different Location

```python
names = ["train A test A", "train B test B", "train A test B", "train A&B test B"]

precision_scores = [0.8455, 0.7864, 0.7579, 0.8359]
f1_scores = [0.8273, 0.7818, 0.8596, 0.8128]
recall_scores = [0.8227, 0.7773, 0.9930, 0.7949]

precision_positions = [0.75, 1.75, 2.75, 3.75]
f1_positions = [1, 2, 3, 4]
recall_positions = [1.25, 2.25, 3.25, 4.25]

plt.bar(precision_positions, precision_scores, width=0.25, color='navy', label = 'precision')
plt.bar(f1_positions, f1_scores, width=0.25, color='mediumblue', label = 'f1')
plt.bar(recall_positions, recall_scores, width=0.25, color='lightblue', label = 'recall')

plt.ylim(0, 1.2)
plt.legend(loc = 'upper right')
plt.title("Different Locations")
plt.xlabel("Train and Test Combinations")
plt.ylabel("Metric Score")
plt.xticks(f1_positions, names)

# Turn on the grid
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='black')
# Customize the minor grid
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

plt.show()
```

## Different Resolution

```python
names = ["train A test A", "train C test C", "train A test C", "train A&C test C"]

precision_scores = [0.8455, 0.7878, 0.3106, 0.85]
f1_scores = [0.8273, 0.8082, 0.4298, 0.8583]
recall_scores = [0.8227, 0.8857, 0.8596, 0.8792]

precision_positions = [0.75, 1.75, 2.75, 3.75]
f1_positions = [1, 2, 3, 4]
recall_positions = [1.25, 2.25, 3.25, 4.25]

plt.bar(precision_positions, precision_scores, width=0.25, color='red', label = 'precision')
plt.bar(f1_positions, f1_scores, width=0.25, color='orange', label = 'f1')
plt.bar(recall_positions, recall_scores, width=0.25, color='yellow', label = 'recall')

plt.ylim(0, 1.2)
plt.legend(loc = 'upper right')
plt.title("Different Resolutions")
plt.xlabel("Train and Test Combinations")
plt.ylabel("Metric Score")
plt.xticks(f1_positions, names)

# Turn on the grid
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='black')
# Customize the minor grid
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

plt.show()
```
